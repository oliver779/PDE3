## Exercise 1

Consider the the function
$$
u(x,y) = x^4 - 6x^2y^2 + y^4
$$

### Part a)

Verify that $u(x,y)$ satisfies the Laplace equation.

```python
import sympy as sp
x,y = sp.symbols('x y')
u = x**4-6*x**2*y**2+y**4
dx = sp.diff(u,x)
dx2 = sp.diff(dx,x)
dy = sp.diff(u,y)
dy2 = sp.diff(dy,y)
print(f'{dx2}  and {dy2}')
if dx2+dy2 ==0:
    print(f'The function {u} satisfies the elliptical Laplacian Partial Differential Equation')
```

### Part b)

- Check the [numpy.meshgrid](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html) documentation
